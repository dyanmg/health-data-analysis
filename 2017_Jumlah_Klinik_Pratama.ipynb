```python
import tabula.io as tb
import pandas as pd
import numpy as np
import math
```

```python
file = 'Profil-Kesehatan-Indonesia-tahun-2017.pdf'
# file = 'https://github.com/dyanmg/health-data-analysis/raw/master/Profil-Kesehatan-Indonesia-Tahun-2020.pdf'
data = tb.read_pdf(file, area = (90, 30, 430, 570), pages = '76')
df = data[0]
df = df.iloc[1:,[0,2]]
df
```

```python
concat = df.copy()
concat.columns = np.arange(2)
split = concat.iloc[:,0].str.split(' ')
isna = concat.iloc[:,1].isna()
concat.loc[isna, 1] = split.str[-1]
concat.loc[isna, 0] = split.str[:-1].str.join(' ') 
concat
```

```python
final = concat.copy()
# final[1] = final[1].astype(np.double)
final[1] = final[1].astype(np.int_)
final.columns = ['provinsi', 'jumlah']
# final.reset_index(drop=True, inplace=True)
final.set_index('provinsi', inplace=True)
final
```

```python
final.to_csv('output/2017_jumlah_klinik_pratama.csv')
```
