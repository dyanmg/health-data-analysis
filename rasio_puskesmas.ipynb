```python
import tabula as tb
import pandas as pd
import numpy as np
import math
```

```python
file = 'Profil-Kesehatan-Indonesia-Tahun-2020.pdf'
# file = 'https://github.com/dyanmg/health-data-analysis/raw/master/Profil-Kesehatan-Indonesia-Tahun-2020.pdf'
data = tb.read_pdf(file, area = (90, 0, 412, 510), pages = '54')
df = data[0]
df.columns = np.arange(2)
jakarta = df.iloc[2,:]
df = df.iloc[3:]
df
```

```python
# df[2] = df[0].str.split(' ')
# df[2] = df[0].str.split(' ').str[-1].str.replace(',', '.')
process = df.copy()
process[1] = process[0].str.split(' ').str[-1]
process[0] = process[0].str.split(' ').str[:-1].str.join(' ')
process
```

```python
concat = pd.concat([jakarta.to_frame().T, process])
concat
```

```python
final = concat.copy()
final[1] = final[1].str.replace(',', '.', regex=False).astype(np.double)
final.columns = ['provinsi', 'rasio']
final
```

```python
final.to_csv('output/rasio_puskesmas.csv', index=False)
```
