```python
import tabula.io as tb
import pandas as pd
import numpy as np
import math
```

```python
file = 'Profil-Kesehatan-Indonesia-tahun-2017.pdf'
# file = 'https://github.com/dyanmg/health-data-analysis/raw/master/Profil-Kesehatan-Indonesia-Tahun-2020.pdf'
data = tb.read_pdf(file, area = (0, 30, 530, 570), pages = '93')
df = data[0]
# df = df.iloc[1:,[0,2]]
data_list = df.columns[0].split('\r')[:-1]
# data_list = [x.replace(',','.') for x in data_list]
data_list
# df
```

```python
num_part = [(''.join(x for x in data_item if x.isnumeric() or x == ',')) for data_item in data_list]
num_part
```

```python
pairs = tuple((x.replace(str(y), ''), float(y.replace(',','.'))) for x, y in zip(data_list, num_part))
pairs
```

```python
df = pd.DataFrame(pairs)
df
```

```python
final = df.copy()
# final[1] = final[1].astype(np.double)
# final[1] = final[1].astype(np.int_)
final.columns = ['provinsi', 'persentase']
# final.reset_index(drop=True, inplace=True)
final.set_index('provinsi', inplace=True)
final
```

```python
final.to_csv('output/2017_persentase_posyandu_aktif.csv')
```
