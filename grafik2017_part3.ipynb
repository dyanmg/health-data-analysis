```python
import tabula.io as tb
import pandas as pd
import numpy as np
import math
```

```python
file = 'Profil-Kesehatan-Indonesia-tahun-2017.pdf'
# file = 'https://github.com/dyanmg/health-data-analysis/raw/master/Profil-Kesehatan-Indonesia-Tahun-2020.pdf'
data = tb.read_pdf(file, area = (0, 50, 150, 700), pages = '10', multiple_tables = False)
data = data[0]
data
```

```python
data.columns = np.arange(6)
data
```

```python
joined = data.loc[:,1:].apply(lambda x: ' '.join(x.dropna()), axis=1)
# data
added = data.loc[:,(0,)]
added.loc[:,1] = joined
added
# joined
```

```python
added.columns = np.arange(2)
added.reset_index(drop=True, inplace=True)
added
```

```python
def my_isnan(var):
    return type(var) == float and math.isnan(var)
    
selected = []
carry_texts = []
for index, row in added[::-1].iterrows():
    if my_isnan(row[0]):
        carry_texts.insert(0, row[1])
    elif len(carry_texts) > 0:
        carry_texts.insert(0, row[1])
        row[1] = ' '.join(carry_texts)
        carry_texts = []
        selected.insert(0, row)
    else:
        selected.insert(0, row)
filtered = pd.DataFrame(selected)
filtered
```

```python
new_index = filtered.reset_index(drop=True)
new_index[0] = new_index[0].str.title()
new_index[1] = new_index[1].str.title()
# new_index[2] = new_index[2].astype(np.int_)
new_index
```

```python
new_index.columns = ['gambar', 'judul']
final = new_index.set_index('gambar')
final
```

```python
part1 = pd.read_csv('output/grafik2017_part1.csv', index_col=0)
part1
```

```python
part2 = pd.read_csv('output/grafik2017_part2.csv', index_col=0)
part2
```

```python
full = pd.concat((part1, part2, final))
full
```

```python
full.to_csv('output/grafik2017_full.csv', sep=';')
```
