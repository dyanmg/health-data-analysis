```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import re
%matplotlib inline
```

## Generate Data

```python
#gabung semua kolom file csv 2017
dirs = [dir for dir in os.listdir('output') if dir.startswith('2017_')]
dfs = [pd.read_csv('output/' + dir, index_col=0) for dir in dirs]
df = pd.concat(dfs, axis=1)
columns = [re.sub(r'(2017_|.csv)', '', col).replace('_', ' ') for col in dirs]
df.columns = columns
df = df.drop(df.columns[[1]], axis=1)
df
```

```python
df = df.set_index('provinsi')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],cmap='greens')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=df.columns)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='Greens', annot=True)
```
