```python
import tabula as tb
import pandas as pd
import numpy as np
import math
```

```python
file = 'Profil-Kesehatan-Indonesia-Tahun-2020.pdf'
# file = 'https://github.com/dyanmg/health-data-analysis/raw/master/Profil-Kesehatan-Indonesia-Tahun-2020.pdf'
data = tb.read_pdf(file, area = (400, 0, 733, 493), pages = '69')
df = data[0]
df.columns = np.arange(2)
jabar = df.iloc[1,:]
df = df.iloc[2:]
df
```

```python
process = df.copy()
process[1] = process[0].str.split(' ').str[-1]
process[0] = process[0].str.split(' ').str[:-1].str.join(' ')
process
```

```python
concat = pd.concat([jabar.to_frame().T, process])
concat
```

```python
final = concat.copy()
final[1] = final[1].astype(np.int_)
final.columns = ['provinsi', 'jumlah']
final.set_index('provinsi', inplace=True)
final
```

```python
final.to_csv('output/jumlah_lab.csv')
```
