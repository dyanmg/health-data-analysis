```python
import tabula.io as tb
import pandas as pd
import numpy as np
import math
```

```python
file = 'Profil-Kesehatan-Indonesia-tahun-2017.pdf'
# file = 'https://github.com/dyanmg/health-data-analysis/raw/master/Profil-Kesehatan-Indonesia-Tahun-2020.pdf'
data = tb.read_pdf(file, area = (220, 30, 590, 570), pages = '69')
df = data[0]
df = df[1:]
df
```

```python
concat = df.copy()
split = concat.iloc[:,0].str.split(' ')
# isna = concat[1].isna()
concat.loc[:, 1] = split.str[-1]
concat.iloc[:, 0] = split.str[:-1].str.join(' ') 
# concat.columns = np.arange(2)
concat
```

```python
final = concat.copy()
final[1] = final[1].str.replace('.', '', regex=False).astype(np.int_)
final.columns = ['provinsi', 'jumlah']
# final.reset_index(drop=True, inplace=True)
final.set_index('provinsi', inplace=True)
final
```

```python
final.to_csv('output/2017_jumlah_puskesmas_kesehatan_kerja_olah_raga_sd.csv')
```
